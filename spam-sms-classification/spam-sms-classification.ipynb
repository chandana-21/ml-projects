```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/chandana-21/ml-projects/main/spam-sms-classification/data.csv?token=ATBJOGFCXEX6KWN4C25IG3DBZWXL6")
```

```python
pd.set_option('display.max_colwidth', None)
df.head()
```

```python
df.groupby("Category").describe()
```

```python
df['spam'] = df.Category.apply(lambda x : 1 if x=='spam' else 0)
df.head()
```

```python
import nltk
import string
```

```python
df.Message = df.Message.astype(str)
df.Message = df.Message.str.lower()
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('','',string.punctuation))

df['wo_punc'] = df.Message.apply(lambda text: remove_punc(text))
df.head()
```

```python
df = df.drop('Message',axis=1)
df.head()
```

```python
from nltk.corpus import stopwords

stopwords = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stopwords])

df['wo_stopwords'] = df.wo_punc.apply(lambda text: remove_stopwords(text))
df.head()
```

```python
df = df.drop('wo_punc',axis=1)
df.head()
```

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df['stemmed'] = df.wo_stopwords.apply(lambda text: stem_words(text))
df.head()
```

```python
df = df.drop('wo_stopwords',axis=1)
df.head()
```

```python
df['Message'] = df['stemmed']
df = df.drop('stemmed',axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message,df.spam, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
```

```python
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
```

```python
new_emails = []
for email in emails:
    new_emails.append(stem_words(remove_stopwords(remove_punc(email))))
    
new_emails
```

```python
emails_count = v.transform(new_emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```
